```python

data = {
    'Year': [2023, 2022, 2021, 2020, 2015, 2010, 2005, 2000],
    'Value': [5936502, 5792683, 5649819, 5501515, 3682691, 1407392, 166179, 6228]
}
df = pd.DataFrame(data)

# Sort the DataFrame by the 'Year' column in ascending order
df = df.sort_values('Year').reset_index(drop=True)

# Create a new DataFrame that includes all years between the minimum and maximum years
full_years = pd.DataFrame({'Year': range(df['Year'].min(), df['Year'].max() + 1)})

# Merge the full range of years with the original data
df_full = pd.merge(full_years, df, on='Year', how='left')

# Fill the missing values using linear interpolation
df_full['Value'] = df_full['Value'].interpolate()
```

```python
from scipy.optimize import curve_fit
```

```python
def model(t, f, b, g):
    return f * np.exp(b * t) - g

t=new_data['Period']
y_observed=new_data['FactValueNumeric']
# Ajustement des paramètres en utilisant curve_fit
initial_guess = [1e6, 0.001, 1e6]  # Valeurs initiales pour f, b, et g
params, covariance = curve_fit(model, t, y_observed, p0=initial_guess,maxfev=100000)

# Extraction des paramètres optimaux
f_opt, b_opt, g_opt = params

print(f"Paramètres optimaux : f = {f_opt}, b = {b_opt}, g = {g_opt}")

# Visualisation de l'ajustement
t_fit = np.linspace(min(t), max(t), 100)
y_fit = model(t_fit, f_opt, b_opt, g_opt)

plt.scatter(t, y_observed, label='Données observées', color='red')
plt.plot(t_fit, y_fit, label='Ajustement du modèle', color='blue')
plt.xlabel('Année')
plt.ylabel('Valeur observée')
plt.legend()
plt.show()
```
